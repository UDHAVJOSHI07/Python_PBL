```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler

df= pd.read_csv("heart (1).csv")

#DATA CLEANING
#handling null values:

df.isnull().sum()

df['Cholesterol'].fillna(df['Cholesterol'].mean(), inplace= True)
df['RestingBP'].fillna(df['RestingBP'].median(), inplace=True)

#handling duplicate values:

df.duplicated().sum()
df.drop_duplicates(inplace=True)

#handling outliers

Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)
IQR = Q3 - Q1
df = df[(df['Age'] >= Q1 - 1.5*IQR) & (df['Age'] <= Q3 + 1.5*IQR)]

#fixing inconsistent data

df['Sex'] = df['Sex'].str.lower().str.strip()

#handling incorrect datatype

df['Age'] = df['Age'].astype(int)

#encoding categorical data

le = LabelEncoder()
df['ExerciseAngina'] = le.fit_transform(df['ExerciseAngina'])
df['Sex'] = le.fit_transform(df['Sex'])

#DATA SCALING
#NORMALIZATION

scaler = MinMaxScaler()
df[['Age','Cholesterol','MaxHR','RestingBP']] =scaler.fit_transform(df[['Age','Cholesterol','MaxHR','RestingBP']])

#converting to new csv

df.to_csv("HealthDiseas_cleaaned.csv",sep=',',index=False)
print("code run succesfully")
df

#FEATURE SELECTION
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X = df.drop("HeartDisease", axis=1)
y = df["HeartDisease"]

# split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# apply forward selection
fs = SequentialFeatureSelector(
    LinearRegression(),
    n_features_to_select=5,
    direction='forward'
)

fs.fit(X_train, y_train)
print("selected features:")
print(X.columns[fs.get_support()])

#apply backword feature elemination
bs = SequentialFeatureSelector(
    LinearRegression(),
    n_features_to_select=5,
    direction='backward'
)

bs.fit(X_train, y_train)
print("selected features:")
print(X.columns[bs.get_support()])
print("\nData:")
df
```
